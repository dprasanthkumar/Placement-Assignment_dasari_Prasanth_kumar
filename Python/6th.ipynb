```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Pokemon data from the Excel file
df = pd.read_excel(r'C:\Users\dpras\OneDrive\Desktop\iNueron\Placement Assignment_dasari_Prasanth_kumar\Python\3_pokemon_data.xlsx')

```

```python
df
```

```python
# Get all Pokemons whose spawn rate is less than 5%
spawn_rate_lt_5 = df[df['spawn_chance'] < 5]
```

```python
# Get all Pokemons that have less than 4 weaknesses
weakness_lt_4 = df[df['weaknesses'].str.count(',') < 3]
```

```python
# Get all Pokemons that have no multipliers at all
no_multipliers = df[df['multipliers'].isna()]
```

```python
# Get all Pokemons that do not have more than 2 evolutions
evolution_lt_2 = df[df['next_evolution'].str.count(',') < 2]
```

```python
# Convert the spawn time to minutes and seconds
df['spawn_time'] = pd.to_datetime(df['spawn_time'], format='%M:%S')
```

```python
df['spawn_time_minutes'] = df['spawn_time'].dt.minute * 60 + df['spawn_time'].dt.second
```

```python
# Get all Pokemons whose spawn time is less than 300 seconds (5 minutes)
spawn_time_lt_300 = df[df['spawn_time_minutes'] < 300]
```

```python
# Get all Pokemon who have more than two types of capabilities
more_than_2_types = df[df['type'].str.count(',') > 1]
```

```python
# Plotting the analysis results

# Bar plot for Pokemons whose spawn rate is less than 5%
plt.figure(figsize=(25, 20))
spawn_rate_lt_5['name'].value_counts().plot(kind='bar', color='blue')
plt.xlabel('Pokemon Name')
plt.ylabel('Count')
plt.title('Pokemons with Spawn Rate Less than 5%')
plt.show()
```

```python
# Bar plot for Pokemons that have less than 4 weaknesses
plt.figure(figsize=(25, 20))
weakness_lt_4['name'].value_counts().plot(kind='bar', color='green')
plt.xlabel('Pokemon Name')
plt.ylabel('Count')
plt.title('Pokemons with Less than 4 Weaknesses')
plt.show()
```

```python
# Bar plot for Pokemons that have no multipliers at all
plt.figure(figsize=(20, 6))
no_multipliers['name'].value_counts().plot(kind='bar', color='orange')
plt.xlabel('Pokemon Name')
plt.ylabel('Count')
plt.title('Pokemons with No Multipliers')
plt.show()
```

```python
# Bar plot for Pokemons that do not have more than 2 evolutions
plt.figure(figsize=(8, 6))
evolution_lt_2['name'].value_counts().plot(kind='bar', color='purple')
plt.xlabel('Pokemon Name')
plt.ylabel('Count')
plt.title('Pokemons with Less than 2 Evolutions')
plt.show()
```

```python
# Bar plot for Pokemons whose spawn time is less than 300 seconds
plt.figure(figsize=(20, 6))
spawn_time_lt_300['name'].value_counts().plot(kind='bar', color='red')
plt.xlabel('Pokemon Name')
plt.ylabel('Count')
plt.title('Pokemons with Spawn Time Less than 300 Seconds')
plt.show()
```

```python
# Bar plot for Pokemon who have more than two types of capabilities
if not more_than_2_types.empty:
    plt.figure(figsize=(8, 6))
    more_than_2_types['name'].value_counts().plot(kind='bar', color='brown')
    plt.xlabel('Pokemon Name')
    plt.ylabel('Count')
    plt.title('Pokemon with More Than Two Types of Capabilities')
    plt.show()
else:
    print("No Pokemon found with more than two types of capabilities.")
```
